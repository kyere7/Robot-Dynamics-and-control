- **(1.3) Components and Structure of Robots**
  -robots have links connected by joints to form a kinematic chain
  - revolute joint allows a rotation, and a prismatic joint allows linear relative motion between two links
  - The number of joints determine the degree of freedom (DOF).
  - the difficulty of controling a manipulator increases rapidly with increasing number of links.
  - **workspace** of a manipulator is total volume used by manipulator as it executes all possible motions.
  - robots can be classified by:
    - **power source:** electrically, hydraulically, or pneumatically powered.
    - **application area:** assembly and non-assembly robots
    - **method of control:** classified by control method into servo and non-servo robots
    - **Geometry:** articulate (RRR), spherical (RRP), SCARA
(RRP), cylindrical (RPP), or cartesian (PPP).  
    *a sixth type is parallel manipulator*
  - **Robotic systems:** A robot manipulator should be viewed as more than just a series of mechanical linkage
  - **Accuracy and repeatability:** accuracy is how close the manipulator can come to a given point within the workspace, while repeatability is the measure of how close a manipulator can return to a previously taught point
